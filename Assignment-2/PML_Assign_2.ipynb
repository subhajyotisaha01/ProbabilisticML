# Part 2: Visualizing GP Priors and Posteriors for Regression (20 marks)

```python
x = np.linspace(0, 4*3.14, 100).reshape(1, 100)
y = np.sin(x) + np.random.normal(0, np.sqrt(0.05))
list_l = np.array([0.2, 0.5, 1, 2, 10])
```

```python
sigma_sq = 0.05
ro_sq = 1
```

```python
for l in list_l:
    
    K = ro_sq * np.exp( (-1) * np.square(np.transpose(x) - x) / l**2)
    small_positive_matrix = 10**(-20) *np.eye(100)
    K = K + small_positive_matrix
    f = np.random.multivariate_normal(np.zeros(100), K, 1).reshape(1, 100)
    f_pos_mean = (1/sigma_sq) * np.dot(y, np.linalg.inv((1/sigma_sq) * np.eye(100) + np.linalg.inv(K) ) )
    # plt.figure(figsize=(10, 10))
    plt.plot(np.transpose(x), np.transpose(f), label = 'GP Prior')
    plt.plot(np.transpose(x), np.transpose(f_pos_mean), label = 'GP Pos Mean')
    plt.plot(np.transpose(x), np.transpose(np.sin(x)), label='True Val')
    plt.title('For l = ' + str(l))
    plt.legend()
    plt.show()
```
